# Reanalysis comparison with observations for long-term climate analysis

Production date 08-2025

Produced by: CNR-ISMAR

## üåç Use case: Evaluating the physical consistency of air-sea heat fluxes for long-term climate analysis

## ‚ùì Quality assessment question

* **Do modern global reanalysis products correctly represent the heat budget closure in the Mediterranean Sea?**

The surface heat budget drives the Mediterranean Sea circulation and moderates the climate of the entire region. The exchange of heat between the atmosphere and the ocean drives the sea's dynamics, primarily by creating dense water. In winter, intense cooling and evaporation in key areas make the surface water colder and saltier, causing it to sink to greater depths. This process, known as deep convection, powers the Mediterranean's deep-ocean circulation, which is vital for transporting oxygen and nutrients.
Furthermore, the sea acts as a massive heat reservoir, absorbing solar energy in the summer and releasing it in the winter, which moderates the regional climate. An accurate surface heat budget is therefore critical for understanding ocean circulation, for accurate weather forecasting, and for predicting how the regional climate may change.
This report evaluates the suitability of ERA5, the fifth-generation atmospheric reanalysis from the European Centre for Medium-Range Weather Forecasts (ECMWF), for this purpose. The assessment is conducted by reviewing and synthesising findings from recent scientific literature that utilise and compare ERA5 data against its predecessor (ERA-Interim), in-situ observations, and higher-resolution atmospheric models.

## üì¢ Quality assessment statements

```{admonition} These are the key outcomes of this assessment

* ERA5 performs well in representing those meteorogical variables used in the bul formulas, better than its predecessor ERA-Interim;

* For the problem of calculating the long-term, basin-wide net heat budget, the standard ERA5 product is limited by its spatial resolution, which causes inability to fully capture extreme winter heat loss events in key regions, leading to a significant uncertainty and results in a spurious net positive heat flux for the Mediterranean basin.
```

## üìã Methodology

The net surface heat flux $Q_{net}$ is defined as the sum of four components: net shortwave radiation $Q_{SW}$, net longwave radiation $Q_{LW}$, latent heat flux $Q_{LH}$, and sensible heat flux $Q_{SH}$. A core concept is the Mediterranean's heat budget closure hypothesis, which posits that to maintain a long-term thermal balance, the basin must lose a net amount of heat at its surface to compensate for the heat gained via the inflow of warmer Atlantic water. A physically consistent analysis should therefore result in a negative basin-averaged $Q_{net}$ [[1]](https://archimer.ifremer.fr/doc/00122/23326/).

These fluxes are calculated using bulk formulas that rely on foundational atmospheric variables extracted from the ERA5 dataset, including 10-meter wind components (zonal and meridional), 2-meter air temperature, dew point temperature, mean sea level pressure and total cloud coverage.

To evaluate the accuracy and performance of ERA5, the Regional Ocean Modelling System (ROMS) was forced separately by ERA5 and its predecessor, ERA-Interim. The resulting sea temperature and salinity were then compared against each other and against observations to quantify the improvement offered by ERA5 [[2]](https://doi.org/10.1016/j.dynatmoce.2024.101525). Reanalysis data are also directly compared with observations. This includes vertical ocean profiles from Argo profilers [[2]](https://doi.org/10.1016/j.dynatmoce.2024.101525) and high-quality, in-situ measurements from the ENEA Oceanographic Observatory near the island of Lampedusa in the Central Mediterranean Sea [[3]](https://doi.org/10.3390/rs13112188). Finally, the basin-wide heat budget calculated from the standard 0.25¬∞ ERA5 is compared against the result from a higher-resolution (0.125¬∞) ECMWF operational analysis to isolate and assess the impact of spatial resolution on the result's physical consistency [[4]](https://doi.org/10.5194/egusphere-2025-2867).

Results are shown for:

**[](reanalysis_reanalysis-era5-single-levels_resolution_q03:reanalysis-era5-single-levels_validation_q04:section-1)**
 * Comparison with ERA-Interim
 * Validation Against In-Situ Data

**[](reanalysis_reanalysis-era5-single-levels_resolution_q03:reanalysis-era5-single-levels_validation_q04:section-2)**

## üìà Analysis and results

(reanalysis_reanalysis-era5-single-levels_resolution_q03:reanalysis-era5-single-levels_validation_q04:section-1)=
### 1. Performance of ERA5 in a regional context

#### Comparison with ERA-Interim

```{figure} 47527289-5bdf-476c-9477-97346feb45f9.jpg
---
height: 550px
---
A comparison between model results using ERA-I and ERA5 with observation data for temperature. The location of the stations can be found on the title of each profile. All 12 stations are for 2018. Reproduced from [[2]](https://doi.org/10.1016/j.dynatmoce.2024.101525).
```

As discussed in the paper, the improvements in the ocean model are a direct result of differences in the atmospheric parameters provided by the two reanalyses: the warmer water in the ERA5 simulation is attributed to a combination of higher shortwave radiation flux and higher air temperatures; the higher salinity is primarily caused by a more negative latent heat flux in ERA5, which signifies greater evaporation, and stronger wind speeds that enhance the vertical mixing of saltier, deeper water to the surface.

To assess the quality of ERA5 for heat budget studies, Marullo et al. (2021) conducted a direct validation of its outputs against a complete annual cycle of in situ measurements. They compared ERA5-derived meteorological variables and radiative fluxes (shortwave and longwave irradiance) with in situ observations. Hourly data from June 2017 to June 2018 relative to the closest grid point to the Lampedusa mooring were used. 

Results show that ERA5's foundational meteorological variables (air temperature, sea surface temperature, pressure, humidity) show a very good agreement with the direct buoy measurements, with very small biases. The main discrepancy noted was a tendency for ERA5 to underestimate wind speed during high-wind conditions. However, when evaluating the radiative flux components, which are essential for the heat budget, the performance was mixed. For shortwave irradiance, ERA5 showed a negligible positive bias of about +1.0 W/m¬≤; for longwave irradiance, ERA5 exhibited a large negative bias of -17 W/m¬≤, indicating that it significantly underestimated the downward longwave radiation compared to in-situ observations.

### 2. Benchmarking against higher-resolution models

While ERA5 shows clear strengths, its application for calculating the long-term net heat budget for the Mediterranean basin reveals a significant limitation tied to its spatial resolution. According to the heat budget closure hypothesis, the Mediterranean Sea must have a net surface heat loss to balance the thermal energy it gains from the Atlantic Ocean inflow. Ghani et al. 2025 [[4]](https://doi.org/10.5194/egusphere-2025-2867) used two atmospheric datasets with different resolutions to calculate the net heat flux over the Mediterranean Sea. To ensure consistency between the hourly ERA5 data and the six-hourly ECMWF analysis [[5]](https://doi.org/10.1002/qj.49712656415), the authors processed the fields from both datasets into daily mean values. 

Using the standard ERA5 reanalysis (0.25¬∞ resolution), it resulted in a basin-averaged net heat flux of $+5 ¬± 1.2 W m^{-2}$. Given that the oceanographic observations at the Strait of Gibraltar indicate an annual-mean loss of $‚àí5.2‚Äâ¬±‚Äâ1.3 W m^{-2}$ [[6]](https://doi.org/10.1029/CE046p0013), this estimate of $Q_{net}$ is largely warm-biased. Previous studies using state-of-the-art reanalyses have shown similar results (see e.g. [[7]](https://doi.org/10.1002/2016JC012254). This positive value suggests a net heat gain, which is physically inconsistent with the closure hypothesis. In contrast, using a high-resolution (0.125¬∞) ECMWF analysis for the same period yielded a net heat flux of $‚àí3.6 ¬± 1.3 W m^{-2}$. This negative value indicates a net heat loss and is consistent with the physical understanding of the basin.

This discrepancy is due to the coarser spatial resolution of ERA5, which is insufficient to accurately capture the intense, localised, and extreme heat loss events that occur during winter in key areas such as the Gulf of Lion and the Aegean Sea. These extreme events are critical in determining the overall negative sign of the Mediterranean's net heat budget. This finding highlights that for the specific problem of basin-wide budget closure, the resolution of the standard ERA5 product is a major limitation.
