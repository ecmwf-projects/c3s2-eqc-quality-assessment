# Reanalysis comparison with observations for long-term climate analysis

Production date 08-2025

Produced by: CNR-ISMAR

## üåç Use case: Evaluating the physical consistency of air-sea heat fluxes for long-term climate analysis

## ‚ùì Quality assessment question

* **Do modern global reanalysis products correctly represent the heat budget closure in the Mediterranean Sea?**

Understanding the heat balance in the Mediterranean Sea has always been a challenging goal, pursued both through numerical models and through the analysis of observational data. The reason for this effort lies in the fact that the Mediterranean is one of the few regions of the world's oceans where the advective heat flux is known with sufficient accuracy to allow verification of datasets and formulas for surface heat flux, and an accurate surface heat balance, i.e. accurate fluxes at the air‚Äìsea interface, are essential for understanding ocean circulation, for accurate weather forecasting, and for predicting how the regional climate might change. 

Measurements taken in the Strait of Gibraltar have consistently indicated a heat flux into the Mediterranean equivalent to an average surface heat loss of less than $10\ W m^{-2}$, but the results obtained have varied depending on the data and parameterisation used for the four components of the net heat flux. The table below shows some results from the literature on this subject.

Here, we assess the suitability of ERA5, the fifth-generation atmospheric reanalysis from the European Centre for Medium-Range Weather Forecasts (ECMWF), in terms of representing the atmospheric variables involved in the bulk formulas and the fluxes themselves, drawing on results from recent scientific literature that uses and compares ERA5 data with those of its predecessor (ERA-Interim), in situ observations, and high-resolution atmospheric models.

````````````


## üì¢ Quality assessment statements

```{admonition} These are the key outcomes of this assessment
:class: note
* ERA5 reproduces near-surface temperature, humidity and pressure with high fidelity, while exhibiting a systematic underestimation of wind speed during strong-wind conditions;
* Its longwave component need to be used carefully since the reanalysis tends to underestimate when compared with observations;
* ERA5 products lead to a spurious positive net heat flux for the Mediterranean basin, possibly due to its spatial resolution and its representation of clouds.
```

## üìã Methodology

The net surface heat flux $Q_{net}$ is defined as the sum of four components: net shortwave radiation $Q_{SW}$, net longwave radiation $Q_{LW}$, latent heat flux $Q_{LH}$, and sensible heat flux $Q_{SH}$. 

$$
Q_{net}= Q_{SW} + Q_{LW} + Q_{LH} + Q_{SH}
$$

A core concept is the Mediterranean's heat budget closure hypothesis, which posits that to maintain a long-term thermal balance, the basin must lose a net amount of heat at its surface to compensate for the heat gained via the inflow of warmer Atlantic water. A physically consistent analysis should therefore result in a negative basin-averaged $Q_{net}$ [[1]](https://archimer.ifremer.fr/doc/00122/23326/), [[2]](https://doi.org/10.1016/j.pocean.2017.07.001).

Direct measurements of turbulent fluxes (i.e. latent and sensible heat fluxes) over the ocean are particularly rare given the difficulties involved in collecting them, and the most common approach, even in numerical climate modelling, is to use empirical methods to estimate the various components of air-sea heat exchange by linking turbulent transfer on a microscale to more easily measurable macroscopic quantities, such as near-surface wind intensity, humidity or temperature. Cloud cover is then included in the estimation of shortwave and longwave radiation components. To evaluate the accuracy and performance of ERA5 both for those atmospheric variables involved and the fluxes, validation against observations is carried with Argo data [[3]](https://argo.ucsd.edu/) and measurements from the ENEA Oceanographic Observatory [[4]](https://www.lampedusa.enea.it/). Finally, the basin-wide heat budget calculated from the standard 0.25¬∞ ERA5 is compared against the result from a higher-resolution (0.125¬∞) ECMWF operational analysis.

Results are shown for:

**[](reanalysis_reanalysis-era5-single-levels_resolution_q03:reanalysis-era5-single-levels_resolution_q04:section-1)**

**[](reanalysis_reanalysis-era5-single-levels_resolution_q03:reanalysis-era5-single-levels_resolution_q04:section-2)**

## üìà Analysis and results

(reanalysis_reanalysis-era5-single-levels_resolution_q03:reanalysis-era5-single-levels_resolution_q04:section-1)=
### 1. Atmospheric forcing fields 

To assess the quality of ERA5 for heat budget studies, Marullo et al. [[5]](https://doi.org/10.3390/rs13112188) acquired measurements over a complete annual cycle, and compared ERA5-derived meteorological variables and radiative fluxes (shortwave and longwave irradiance) with these in-situ observations. Hourly data from June 2017 to June 2018 relative to the closest grid point to the Lampedusa (Sicily, Italy) mooring were used. The comparison demonstrates that ERA5 reproduces near-surface temperature, humidity and pressure with high fidelity, while exhibiting a systematic underestimation of wind speed during strong-wind conditions. This behaviour is directly relevant for bulk flux computations: weaker winds reduce both latent and sensible heat losses through lower turbulent transfer, thus favouring a more positive net heat flux into the ocean. The study also identifies non-negligible discrepancies in downwelling longwave irradiance, with the ERA5 Mean surface downward long-wave radiation flux displaying a substantial negative bias relative to in situ observations. Errors in longwave fluxes are particularly significant for the Mediterranean Sea, where the radiative term represents a major component of the net surface heat balance [[6]](https://doi.org/10.1029/94JC02496).

```{figure} 7ad92125-8643-4549-8aaf-268663a57274.png
---
height: 550px
---
Comparison between key meteorological variables contributing to air‚Äìsea heat flux estimates from ERA5 and from in situ observations. Air temperature (a), dew point temperature (b), atmospheric pressure at sea level (c), wind intensity (d), sea surface temperature (e). The (hourly) SST was inferred from the ‚Äúskin‚Äù SST by adding the mean value of the difference with the ‚Äúsubskin‚Äù (0.17 ¬∞C) to make it more comparable with the sensor measurement at 1 m depth. Dot color indicates data density, increasing from white to black. Data refer to the period 3 June 2017‚Äì3 June 2018. Boxes within each plot include statistics of differences between ERA5 estimates and in situ measurements. Negative bias values indicate underestimates of the reanalysis. Reproduced from [[5]](https://doi.org/10.3390/rs13112188).
```

```{figure} a284920e-ba9b-4b05-b3fa-f28873dd651c.png
---
height: 300px
---
Comparison between hourly shortwave (a) and longwave (b) irradiance estimated by ERA5 and corresponding values measured on the buoy. Dot color indicates data density, increasing from white to black. Data refer to the period 4 June 2017‚Äì3 June 2018. Units are W/m2. Boxes within each plot include statistics of differences between ERA5 estimates and in situ measurements of shortwave and longwave irradiance. Reproduced from [[5]](https://doi.org/10.3390/rs13112188).
```

The modelling results shown below, reveal physical manifestations of these atmospheric biases. The Regional Ocean Modeling System (ROMS) is a three-dimensional, free-surface, terrain-following hydrodynamic model. In Babagolimatikolaei [[7]](https://doi.org/10.1016/j.dynatmoce.2024.101525), ROMS was configured for the Adriatic Sea - which is the northernmost arm of the Mediterranean Sea - at a horizontal resolution of 1.5‚Äì2.5 km and forced with daily atmospheric fields derived from ERA5 and its predecessor, ERA-Interim (ERA-I). The model runs for 7 years, spanning from 2012 to 2018, and below are shown results for temperature and salinity for each month of 2018. 

The ERA5-forced simulation systematically produces warmer and more saline upper-ocean conditions relative to ERA-Interim, and generally shows a closer agreement with observations in the central and southern Adriatic. The warmer SST is consistent with reduced turbulent heat losses associated with ERA5‚Äôs weaker winds, while the enhanced salinity aligns with decreased evaporation-driven freshwater removal. 

Taken together, the two studies highlight a coherent mechanism linking atmospheric forcing characteristics to ocean model behaviour. Small but systematic deviations in key atmospheric variables‚Äîparticularly wind speed and radiative fluxes‚Äîpropagate through bulk formulae into the surface heat budget, ultimately shaping the modelled thermohaline structure.     

```{figure} 47527289-5bdf-476c-9477-97346feb45f9.jpg
---
height: 550px
---
A comparison between model results using ERA-I and ERA5 with observation data for sea surface temperature. The location of the stations can be found on the title of each profile. All 12 stations are for 2018. Reproduced from [[7]](https://doi.org/10.1016/j.dynatmoce.2024.101525).
```

```{figure} f9a181fc-66e1-4a73-a30a-ad8258ea42af.jpg
---
height: 550px
---
A comparison between model results using ERA-I and ERA5 with observation data for salinity. The location of the stations can be found on the title of each profile. All 12 stations are for 2018. Reproduced from [[7]](https://doi.org/10.1016/j.dynatmoce.2024.101525).

(reanalysis_reanalysis-era5-single-levels_resolution_q03:reanalysis-era5-single-levels_resolution_q04:section-2)=
### 2. Surface heat fluxes balance

Although ERA5 has clear strengths, in Ghani et al. [[8]](https://doi.org/10.5194/egusphere-2025-2867) is shown that it should be used with caution for calculating the long-term budget for the Mediterranean. Using the reanalysis input fields for the 15-year period from 2006 to 2020, it resulted in a basin-averaged net heat flux of $+5 ¬± 1.2 W m^{-2}$. This positive value suggests a net heat gain, which is physically inconsistent with the closure hypothesis.

The authors hypothesise that the uncertainty can be related to the atmospheric forcing spatial resolution, and possibly cloud cover (not shown), and to demonstrate this they use ECMWF analysis [[9]](https://doi.org/10.1002/qj.49712656415) fields as input, which have a resolution of 0.125¬∞ instead of the 0.25¬∞ of the reanalysis. The problem related to the changes undergone in the operational datasets during the averaging period, in terms of model resolution and the assimilated number of observations, is addressed using the dataset with approximately uniform model resolution and physics. Using this high-resolution analysis, yielded a net heat flux of $‚àí3.6 ¬± 1.3 W m^{-2}$, which is now consistent with the closure hypothesis. 

In particular, as shown in the figure below, the coarser spatial resolution of ERA5 could be insufficient to accurately capture the intense, localised, and extreme heat loss events that occur during winter in key areas. The map derived from high-resolution data (top panel) clearly shows intense net heat loss (deep blue colors) in critical regions like the Gulf of Lion and the Aegean Sea, while the map from ERA5 (bottom panel), shows significantly weaker heat loss in these same areas and larger regions of net heat gain (yellow/orange colors), possibly leading to an overall warmer balance of the basin.

Consistent with the findings also found in the previous section, similar conclusions on the impact of spatial resolution and cloud representation on the quality of the ERA5 surface longwave radiation, also in other parts besides the Mediterranean, were reached e.g. in Silber et al. (2019) [[10]](https://doi.org/10.1175/JCLI-D-19-0149.1), Lopes et al. (2024) [[11]](https://doi.org/10.1029/2023JD040306), Biagio et al. (2020) [[12]](https://doi.org/10.1029/2020JD032555).  
 

```{figure} 634b0f24-23c8-490d-be74-c6a2d709166e.png
---
height: 450px
---
Comparison of the annual $Q_{net}$(W m‚Åª¬≤) computed from, a) ECMWF and b) ERA-5 input datasets. Reproduced from [[8]](https://doi.org/10.5194/egusphere-2025-2867).
```

## ‚ÑπÔ∏è If you want to know more	

[Through a Glass Darkly: Envisioning ocean heat budget](https://www.earthdata.nasa.gov/news/feature-articles/through-glass-darkly-envisioning-ocean-heat-budget) 

[Closing the sea surface mixed layer temperature budget from in situ observations alone: Operation Advection during BoBBLE](https://doi.org/10.1038/s41598-020-63320-0)

[Long term evolution of heat budget in the Mediterranean Sea from Med-CORDEX forced and coupled simulations](https://doi.org/10.1007/s00382-016-3363-5)

### Key resources

Some key resources and further readings were linked throughout this assessment. 

<a id='The_CDS_catalogue_entries_for_the_data_used_were:'></a>
The CDS catalogue entries for the data used were:

* ERA5 hourly data on single levels from 1940 to present: https://cds.climate.copernicus.eu/datasets/reanalysis-era5-single-levels?tab=overview

* ERA5 monthly averaged data on single levels from 1940 to present: https://cds.climate.copernicus.eu/datasets/reanalysis-era5-single-levels-monthly-means?tab=overview

### References

[[1]](https://archimer.ifremer.fr/doc/00122/23326/) Bethoux, J. P. (1979). Budgets of the Mediterranean Sea: their dependence on the local climate and on the characteristics of the Atlantic waters. Oceanologica Acta, 2(2), 157‚Äì163.

[[2]](https://doi.org/10.1016/j.pocean.2017.07.001) G. Jord√†, K. Von Schuckmann, S.A. Josey, G. Caniaux, J. Garc√≠a-Lafuente, S. Sammartino, E. √ñzsoy, J. Polcher, G. Notarstefano, P.-M. Poulain, F. Adloff, J. Salat, C. Naranjo, K. Schroeder, J. Chiggiato, G. Sannino, D. Mac√≠as, The Mediterranean Sea heat and mass budgets: Estimates, uncertainties and perspectives, Progress in Oceanography,
Volume 156, 2017, Pages 174-208, ISSN 0079-6611.

[[3]](https://argo.ucsd.edu/) Argo: an international program that measures water properties across the world‚Äôs ocean.

[[4]](https://www.lampedusa.enea.it/) ENEA station for climate observations.

[[5]](https://doi.org/10.3390/rs13112188) Marullo, S.; Pitarch, J.; Bellacicco, M.; Sarra, A.G.d.; Meloni, D.; Monteleone, F.; Sferlazzo, D.; Artale, V.; Santoleri, R. Air‚ÄìSea Interaction in the Central Mediterranean Sea: Assessment of Reanalysis and Satellite Observations. Remote Sens. 2021, 13, 2188. 

[[6]](https://doi.org/10.1029/94JC02496) Bignami, F., S. Marullo, R. Santoleri, and M. E. Schiano (1995), Long wave radiation budget on the Mediterranean Sea, J. Geophys. Res., 100, 2501‚Äì251.

[[7]](https://doi.org/10.1016/j.dynatmoce.2024.101525) Babagolimatikolaei, J. (2025). A Comparative Study of the Sensitivity of an Ocean Model Outputs to Atmospheric Forcing: ERA-Interim vs. ERA5 for Adriatic Sea Ocean Modelling. Dynamics of Atmospheres and Ocean.

[[8]](https://doi.org/10.5194/egusphere-2025-2867) Ghani, M. H., Pinardi, N., Navarra, A., Mentaschi, L., Bianconcini, S., Maicu, F., and Trotta, F.: Revisited heat budget and probability distributions of turbulent heat fluxes in the Mediterranean Sea, EGUsphere [preprint], 2025.

[[9]](https://doi.org/10.1002/qj.49712656415) Rabier,  F.,  J√§rvinen,  H.,  Klinker,  E.,  Mahfouf,  J.  F.,  &  Simmons,  A.  (2000).    The  ECMWF  operational implementation  of  four‚Äêdimensional  variational  assimilation.  I:  Experimental  results  with  simplified  physics. Quarterly  Journal  of  the  Royal  Meteorological  Society,  126(564),  1143-1170.

[[10]](https://doi.org/10.1175/JCLI-D-19-0149.1) Silber, I., J. Verlinde, S. Wang, D. H. Bromwich, A. M. Fridlind, M. Cadeddu, E. W. Eloranta, and C. J. Flynn, 2019: Cloud Influence on ERA5 and AMPS Surface Downwelling Longwave Radiation Biases in West Antarctica. J. Climate, 32, 7935‚Äì7949.

[[11]](https://doi.org/10.1029/2023JD040306) Lopes, F. M., Dutra, E., Trigo, I. F., &Wild, M. (2024). Evaluation of downward surface longwave flux estimates using Meteosat cloud observations. Journal of Geophysical Research: Atmospheres, 129,e2023JD040306.

[[12]](https://doi.org/10.1029/2020JD032555) Di Biagio, C., Pelon, J., Blanchard, Y.,Loyer, L., Hudson, S. R., Walden, V. P. et al. (2021). Toward a better surface radiation budget analysis over sea ice inthe high Arctic Ocean: a comparative study between satellite, reanalysis,and local-scale observations. Journal of Geophysical Research: Atmospheres,126, e2020JD032555.
